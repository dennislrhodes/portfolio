## Using DDL, create each of the relations in the postgres server. 

```python
import getpass
mypasswd = getpass.getpass()
username = 'dlr6w3'
host = 'pgsql.dsa.lan'
database = 'dsa_student'
```

```python
##Create connection
from sqlalchemy.engine.url import URL
from sqlalchemy import create_engine

postgres_db = {'drivername': 'postgres',
               'username': username,
               'password': mypasswd,
               'host': host,
               'database' :database}
engine = create_engine(URL(**postgres_db), echo=True)
del mypasswd
```

```python
query = """

DROP TABLE IF EXISTS crime CASCADE;

CREATE TABLE crime(
    crime_id     INT NOT NULL UNIQUE,
    iucr         VARCHAR(10) NOT NULL,
    fbi          VARCHAR(10) NOT NULL,
    primary_desc VARCHAR(50) NOT NULL,
    description  VARCHAR(100) NOT NULL,
    PRIMARY KEY (crime_id)
) 


"""

with engine.connect() as connection:
    res = connection.execute(query)
    print(res)
```

```python
query = """

DROP TABLE IF EXISTS location CASCADE;

CREATE TABLE location (
    location_id             INT NOT NULL UNIQUE,
    location_description    VARCHAR(50) NOT NULL,
    block                   VARCHAR(50) NOT NULL,
    beat                    INT NOT NULL,
    district                INT NOT NULL,
    ward                    FLOAT NOT NULL,
    community_area          FLOAT NOT NULL,
    latitude                FLOAT NOT NULL,
    longitude               FLOAT NOT NULL,
    x_coordinate            FLOAT NOT NULL,
    y_coordinate            FLOAT NOT NULL,
    PRIMARY KEY (location_id)
    
)

"""

with engine.connect() as connection:
    res = connection.execute(query)
    print(res)
```

```python
query = """

DROP TABLE IF EXISTS crime_report CASCADE;

CREATE TABLE crime_report(
    case_id        VARCHAR(12) NOT NULL UNIQUE,
    case_number    VARCHAR(12) NOT NULL,
    crime_id       INT NOT NULL,
    location_id    INT NOT NULL,
    arrest         BOOL NOT NULL,
    domestic       BOOL NOT NULL,
    case_date      TIMESTAMP NOT NULL,
    case_year      INT NOT NULL,
    updated_on     TIMESTAMP NOT NULL,
    PRIMARY KEY (case_id),
    FOREIGN KEY (crime_id) REFERENCES crime(crime_id),
    FOREIGN KEY (location_id) REFERENCES location(location_id)


)

"""

with engine.connect() as connection:
    res = connection.execute(query)
    print(res)
```

## Load the data from the given csv file to the relations


```python
##Load the csv
import pandas as pd
datapath = "/dsa/data/DSA-7030/Chicago-Crime-Sample-2012-updated.csv"
df = pd.read_csv(datapath, index_col=0)
```

```python
df.head()
```

```python
## Look at dimensions and Count NaNs
print(df.shape)
df.isna().sum()
```

```python
##Remove nans with values to not violate NOT NULL
df['Case Number'] = df['Case Number'].where(pd.notnull(df['Case Number']), -1)
df['Location Description'] = df['Location Description'].where(pd.notnull(df['Location Description']), 'NA')
df['Ward'] = df['Ward'].where(pd.notnull(df['Ward']), -1.00)
df['Community Area'] = df['Community Area'].where(pd.notnull(df['Community Area']), -1.00)
df['X Coordinate'] = df['X Coordinate'].where(pd.notnull(df['X Coordinate']), -1.00)
df['Y Coordinate'] = df['Y Coordinate'].where(pd.notnull(df['Y Coordinate']), -1.00)
df['Latitude'] = df['Latitude'].where(pd.notnull(df['Latitude']), -1.00)
df['Longitude'] = df['Longitude'].where(pd.notnull(df['Longitude']), -1.00)
df['Location'] = df['Location'].where(pd.notnull(df['Location']), 'NA')
```

```python
## Look at updated dimensions and check if NaNs are gone
print(df.shape)
df.isna().sum()
```

```python
##Create crime dataframe
crime_df=pd.DataFrame({
                        'IUCR':  df.IUCR,
                        'FBI Code': df['FBI Code'],
                        'Primary Type': df['Primary Type'],
                        'Description':  df.Description,
                    
                    
})
print(crime_df.shape)
crime_df.head()
```

```python
##Get unique rows
crime_df_unique = crime_df.drop_duplicates()
print(crime_df_unique.shape)
crime_df_unique.head()
```

```python
##Create Id column
crime_df_unique['crime_id'] = range(0, len(crime_df_unique.IUCR))
crime_df_unique.tail()
```

```python
##Create location df

location_df=pd.DataFrame({
                        
                'Location Description':   df['Location Description'] ,
                'Block':                  df.Block ,
                'Beat':                    df.Beat,
                'District':               df.District ,
                'Ward':                  df.Ward  ,
                'Community Area':          df['Community Area'],
                'Latitude':               df.Latitude ,
                'Longitude':              df.Longitude ,
                'X Coordinate':            df['X Coordinate'],
                'Y Coordinate':          df['Y Coordinate'] ,  
                    
})
print(location_df.shape)
location_df.head()
```

```python
##Get unique rows
location_df_unique = location_df.drop_duplicates()
print(location_df_unique.shape)
location_df_unique.head()
```

```python
##Create Id column
location_df_unique['Location_ID'] = range(0, len(location_df_unique.Block))
print(location_df_unique.shape)
location_df_unique.tail()
```

```python
##Join data to original df to add location and crime ids
join1 =  pd.merge(df, location_df_unique, how='left', on=['Location Description', 'Block', 'Beat', 'District', 'Ward', 'Community Area', 'Community Area', 'Latitude', 'Longitude', 'X Coordinate', 'Y Coordinate'])
```

```python
print(join1.shape)
join1.tail()
```

```python
join2 = pd.merge(join1, crime_df_unique, how='left', on=['IUCR', 'FBI Code', 'Primary Type', 'Description'])
```

```python
print(join2.shape)
join2.head()
```

```python
##Add case id
join2['case_id'] = range(0, len(join2['Case Number']))
join2.tail()
```

```python
##Create crime_report df

creport_df_sql=pd.DataFrame({
                'case_id': join2.case_id,
                'case_number':  join2['Case Number']  ,
                'crime_id': join2['crime_id'],
                'location_id': join2['Location_ID'],
                'arrest':       join2.Arrest  ,
                'domestic':     join2.Domestic  ,
                'case_date':    join2.Date  ,
                'case_year':     join2.Year ,
                'updated_on':    join2['Updated On'] ,       
                    
                    
})
print(creport_df_sql.shape)
creport_df_sql.head()

```

```python
##change the two date columns to timestamp
creport_df_sql.case_date = pd.to_datetime(creport_df_sql.case_date, infer_datetime_format=True) 
creport_df_sql.updated_on = pd.to_datetime(creport_df_sql.updated_on, infer_datetime_format=True)
```

```python
##check NAs
print(creport_df_sql.isna().sum())
print(location_df_unique.isna().sum())
print(crime_df_unique.isna().sum())
```

```python
##create sql crime df
crime_df_sql = pd.DataFrame({
                        'crime_id':  crime_df_unique['crime_id'],
                        'iucr':  crime_df_unique.IUCR,
                        'fbi': crime_df_unique['FBI Code'],
                        'primary_desc': crime_df_unique['Primary Type'],
                        'description':  crime_df_unique.Description,
                    
                    
})
print(crime_df_sql.shape)
crime_df_sql.head()
```

```python
##create sql location df
location_df_sql = pd.DataFrame({
                        
                'location_id':            location_df_unique['Location_ID'],
                'location_description':   location_df_unique['Location Description'] ,
                'block':                  location_df_unique.Block ,
                'beat':                    location_df_unique.Beat,
                'district':               location_df_unique.District ,
                'ward':                  location_df_unique.Ward  ,
                'community_area':          location_df_unique['Community Area'],
                'latitude':               location_df_unique.Latitude ,
                'longitude':              location_df_unique.Longitude ,
                'x_coordinate':            location_df_unique['X Coordinate'],
                'y_coordinate':          location_df_unique['Y Coordinate'] ,  
                    
})
print(location_df_sql.shape)
location_df_sql.head()
```

```python
##Insert crime table
crime_df_sql.to_sql('crime', # The table to load
          engine,             # The engine created above
          schema= username,   # The schema where the table lives, our pawprint
          if_exists='append', # If the table is found, it would keep loading the end of table.
          index=False,        # Recall that panda data frame has a row index, so we need to ignore it
          chunksize=100)       # Do 100 records from the data frame at a time
```

```python
##Insert location table
location_df_sql.to_sql('location', # The table to load
          engine,             # The engine created above
          schema= username,   # The schema where the table lives, our pawprint
          if_exists='append', # If the table is found, it would keep loading the end of table.
          index=False,        # Recall that panda data frame has a row index, so we need to ignore it
          chunksize=10000)       # Do 10000 records from the data frame at a time
```

```python
creport_df_sql.to_sql('crime_report', # The table to load
          engine,             # The engine created above
          schema= username,   # The schema where the table lives, our pawprint
          if_exists='append', # If the table is found, it would keep loading the end of table.
          index=False,        # Recall that panda data frame has a row index, so we need to ignore it
          chunksize=100000)       # Do 100000 records from the data frame at a time
```
